```python
import pandas as pd
import numpy as np
import json
```

```python
f = open('Outputs/neighbor-districts-modified.json')

data = json.load(f)

f.close()
```

```python
data
```

```python
len(data)
```

```python
df = pd.DataFrame(columns=['node1','node2'])
df
```

```python
list1 = []
for k,v in data.items():
    list1.append(k)
    for neigh in v:
        if neigh not in list1:
            df.loc[len(df.index)] = [k,neigh]
```

```python
df.to_csv('Outputs/edge-graph.csv',index=False)
```
